```python
import pandas as pd
import matplotlib.pyplot as plt

parks = pd.read_csv("fa329e3d-89ff-4708-8ddf-81bfedcad11d.csv")
neighborhood = parks["neighborhood"]
count = neighborhood.value_counts()
plt.figure(figsize=(12,10))
count.plot(kind="bar")
plt.xlabel("Neighborhood")
plt.ylabel("Number of Parks")
plt.title("Parks in Neighborhood")
plt.show()
```
