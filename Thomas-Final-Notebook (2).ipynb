Our metric was what neighborhood would be the best for raising a dog. We originally wanted to do best neighborhood to get rich in but due to a lack of easy to access data we scrapped that idea and shifted due to our shared love of dogs.


```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

df = pd.read_csv("https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02")
neighborhood = df["neighborhood"]
count = neighborhood.value_counts()
plt.figure(figsize = (12,10))
count.plot(kind = 'bar')
plt.xlabel("Neighborhood")
plt.ylabel("Number of Fires")
plt.title("Fires per Neighborhood")
plt.show()
```
